```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("CAFE_SALES_CLEANED(in).csv")
df.head(20)
```

```python
df.shape
```

```python
df.columns.tolist()
```

```python
df.columns = df.columns.str.strip()
df.columns.tolist()
```

```python
df.rename(columns={
    'Transaction number': 'Transaction_Number',
    'Item': 'Item',
    'Quantity_Avg': 'Quantity',
    'Price_per_unit_Avg': 'Price_Per_Unit',
    'Total_spent_Avg': 'Total_Spent',
    'Payment Method': 'Payment_Method',
    'Transaction Date': 'Transaction_Date'
}, inplace=True)

df.columns.tolist()
```

```python
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'])
df['Transaction_Date'].dtype
```

```python
df['Month'] = df['Transaction_Date'].dt.month_name()
df['Year'] = df['Transaction_Date'].dt.year
df.head(10)
```

```python
df.dtypes
```

```python
df.describe()
```

```python
total_revenue = df['Total_Spent'].sum()
print(f"Total Revenue: ${total_revenue:,.2f}")
```

```python
best_selling = df.groupby('Item')['Quantity'].sum().sort_values(ascending=False)
print(best_selling)
```

```python
revenue_by_item = df.groupby('Item')['Total_Spent'].sum().sort_values(ascending=False)
print(revenue_by_item)
```

```python
revenue_by_payment = df.groupby('Payment_Method')['Total_Spent'].sum().sort_values(ascending=False)
print(revenue_by_payment)
```

```python
revenue_by_location = df.groupby('Location')['Total_Spent'].sum().sort_values(ascending=False)
print(revenue_by_location)
```

```python
revenue_by_month = df.groupby('Month')['Total_Spent'].sum().sort_values(ascending=False)
print(revenue_by_month)
```

```python
df.to_csv("CAFE_SALES_FINAL.csv", index=False)
print("File saved successfully!")
```
